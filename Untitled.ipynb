```python
df2 = df.drop(['usage_date', 'creation_date'], axis=1)
```

```python
df2.head()
```

```python
model = KMeans(n_clusters=3, random_state=0).fit(df2)
```

```python
predicted = model.labels_
centroids = model.cluster_centers_

print("Predicted clusters to points: ", predicted)
print("Location of centroids: ")
print(centroids)
```

```python
df2['predicted'] = predicted
df2.head()
```
