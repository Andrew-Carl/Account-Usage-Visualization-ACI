```python
import pandas as pd
import numpy as np

from bokeh.plotting import figure, output_file, show
import seaborn as sns
from matplotlib import pyplot as plt

plt.style.use('ggplot')
sns.set(style="whitegrid")

%matplotlib inline
```

```python
df = pd.read_csv('./account_usage_data.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
# Change column names to make them easier to work with.

df.rename(columns = {'Usage Date':'usage_date', 'Account Creation Date':'creation_date', 
                     'Days Since Account Started':'days_since_account_started', 'Number of Uses':'number_of_uses'},
                      inplace=True)
```

```python
df.dtypes
```

```python
# Remove commas from number of uses column and change into an integer.

df['number_of_uses'] = df['number_of_uses'].str.replace(',','').astype('int64')
```

```python
df.dtypes
```

```python
df_uses = df.groupby(df.days_since_account_started, as_index=False).agg({
        'number_of_uses': np.sum,
        'days_since_account_started': lambda x: x.iloc[0]
                }).sort_values(by='number_of_uses', ascending=False)
```

```python
plt.figure(figsize=(20,14))
plt.style.use('fivethirtyeight')
ax = sns.lmplot(x = 'days_since_account_started', y = 'number_of_uses', data = df_uses)
ax.set(ylabel='Number of Uses', 
       xlabel='Days of Account', 
       title='Highest Number of Uses Compared to Days')
```

```python
sns.set(color_codes=True)
for column in ['number_of_uses']:
    plt.figure(figsize=(20,14))
    ax = sns.regplot(x=column, y='days_since_account_started', data=df)
    ax.set(ylabel='Number of Uses', 
       xlabel='Days of Account', 
       title='Highest Number of Uses Compared to Days')
```

```python
sns.set(color_codes=True)
for column in ['days_since_account_started']:
    plt.figure(figsize=(20,14))
    plt.rcParams['axes.labelsize', axes.] = 30
    ax = sns.regplot(x=column, y='number_of_uses', data=df)
    ax.set(ylabel='Number of Uses', 
       xlabel='Days of Account', 
       title='Highest Number of Uses Compared to Days')
```
