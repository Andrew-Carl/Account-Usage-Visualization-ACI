# Linear Regression

```python
x = df['days_since_account_started']
y = df['number_of_uses']


### Plot the data
# create a figure
fig = plt.figure(figsize=(15,7))
# get the axis of that figure
ax = plt.gca()

# plot a scatter plot on it with our data
ax.scatter(x, y, c='k')
ax.plot((df['days_since_account_started'].min(), df['days_since_account_started'].max()),
        (np.mean(df['number_of_uses']), np.mean(df['number_of_uses'])), 
         color='r');
ax.set(ylabel='Number of Uses', 
       xlabel='Days Since Account Started')
```

```python
df['Mean_Yhat'] = y.mean()
```

```python
df.head()
```

```python
df.drop(['usage_date', 'creation_date'], axis=1)
```

```python
# calculate MSE

df['Mean_Yhat_SE'] = np.square(df['number_of_uses'] - df['Mean_Yhat'])
df['Mean_Yhat_SE'].mean()
```

```python
# Calculate beta coefficients

y_bar = df['number_of_uses'].mean()
x_bar = df['days_since_account_started'].mean()
std_y = df['number_of_uses'].std()
std_x = df['days_since_account_started'].std()
r_xy = df.corr().loc['days_since_account_started','number_of_uses']
beta_1 = r_xy*(std_y/std_x)
beta_0 = y_bar - beta_1*x_bar
```

```python
df['Linear_Yhat'] = beta_0 + beta_1*df['days_since_account_started']
```

```python
np.square(df['number_of_uses'] - df['Linear_Yhat']).mean()
```

```python
x = df['days_since_account_started']
y = df['number_of_uses']

# create a figure
fig = plt.figure(figsize=(15,7))
# get the axis of that figure
ax = plt.gca()

# plot a scatter plot on it with our data
ax.scatter(x, y, c='k')
ax.plot(df['days_since_account_started'], df['Linear_Yhat'], color='r');
ax.set(ylabel='Number of Uses', 
       xlabel='Days Since Account Started')
```

```python
# Plot residuals.

# create a figure
fig = plt.figure(figsize=(15,7))
fig.set_figheight(8)
fig.set_figwidth(15)

# get the axis
ax = fig.gca()

# plot it
ax.scatter(x=df['days_since_account_started'], y=df['number_of_uses'], c='k')
ax.plot(df['days_since_account_started'], df['Linear_Yhat'], color='k');

# iterate over predictions
for _, row in df.iterrows():
    plt.plot((row['days_since_account_started'], row['days_since_account_started']), 
             (row['number_of_uses'], row['Linear_Yhat']), 'r-')
    ax.set(ylabel='Number of Uses', 
       xlabel='Days Since Account Started')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df2 = df.drop(['usage_date', 'creation_date'], axis=1)
```

```python
y = df2['number_of_uses']
X = df2.drop(['number_of_uses'], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
print(X.shape, y.shape)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
pred_train = lm.predict(X_train)
pred_test = lm.predict(X_test)
```

```python
# MSE for train data.
np.mean((y_train - lm.predict(X_train)) ** 2)
```

```python
# MSE for test data.
np.mean((y_test - lm.predict(X_test)) ** 2)
```
